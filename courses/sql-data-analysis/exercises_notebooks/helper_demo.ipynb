# Helper Module Demo
Use this notebook as a quick reference for importing `sql_helper` and running queries from any local exercise notebook.

## Steps
1. Import `run_sql` and `get_config` from `sql_helper`.
2. Inspect the resolved Docker network and image to ensure they match your environment.
3. Write your SQL as a string and call `run_sql(your_query, description)` to execute it.

> The helper automatically reads `.env` (if present) to override the Docker network or image.

```python
from sql_helper import run_sql, get_config

config = get_config()
print("Helper configured.\n")
print(f"Network: {config['docker_network']}")
print(f"Image: {config['docker_image']}")
```

```python
query = "SELECT 1 AS sample_result;"
run_sql(query, "Demo query")
```
